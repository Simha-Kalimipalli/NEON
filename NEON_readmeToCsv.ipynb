```python
import time
import math
import csv
import os
import numpy as np
import pandas as pd
```

```python
filename='NEON.D16.ABBY.DP3.30006.001.readme.20220302T205717Z.txt'
df = pd.read_csv(filename, header = None, sep = "\t")
```

```python
#select h5 file name in the readme file
h5file = df.loc[df[0].str.endswith('.h5')]
```

```python
h5file
```

## split h5 file name to extract Xmin and Ymin

```python
h5file_split=h5file[0].str.split('_',expand=True)
```

```python
print (h5file_split)
```

```python
h5file= h5file.join(h5file_split[4]).join(h5file_split[5])
```

```python
h5file
```

```python
h5file_ren=h5file.rename(columns={0: "Filename", 4: "Xmin", 5:"Ymin"})
```

```python
h5file_ren
```

## reset index

```python
h5file_ren = h5file_ren.assign(Index=range(len(h5file_ren))).set_index('Index')
```

```python
h5file_ren
```

 ## convert the column Xmin, Ymin to int

```python
h5file_ren = h5file_ren.astype({"Xmin":"int","Ymin":"int"})
```

```python
 h5file_ren['Xcenter'] =  h5file_ren['Xmin']+500 
```

```python
 h5file_ren['Ycenter'] =  h5file_ren['Ymin']+500 
```

```python
 h5file_ren.to_csv('output.csv')
```
